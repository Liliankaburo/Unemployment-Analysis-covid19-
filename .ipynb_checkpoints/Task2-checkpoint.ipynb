# UNEMPLOYMENT ANALYSIS

# Importing files

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the datasets

```python
# Load the datasets
df1 = pd.read_csv('Unemployment in India.csv')
df2 = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')

# Display the first few rows of each dataset
print("Dataset 1:")
print(df1.head())

print("............................................................................................")

print("\nDataset 2:")
print(df2.head())
```

# Dataset Exploration

```python
print(df1.info())
print("************************************************************************************************")
print(df2.info())
```

```python
print(df1.columns)
print("************************************************************************************************")
print(df2.columns)
```

```python
# Check the shape of the datasets
print("Shape of Dataset 1:", df1.shape)
print("Shape of Dataset 2:", df2.shape)
print("............................................................................................")

# Check for missing values
print("\nMissing values in Dataset 1:")
print(df1.isnull().sum())
print("............................................................................................")

print("\nMissing values in Dataset 2:")
print(df2.isnull().sum())
print("............................................................................................")

# Get basic statistics
print("\nStatistics for Dataset 1:")
print(df1.describe())
print("............................................................................................")

print("\nStatistics for Dataset 2:")
print(df2.describe())
print("............................................................................................")
```

# Data Cleaning

```python
# Drop rows with missing values (if necessary)
df1 = df1.dropna()
df2 = df2.dropna()


# Strip spaces from column names
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# Now convert date columns to datetime format
df1['Date'] = pd.to_datetime(df1['Date'], format='%d-%m-%Y', errors='coerce')
df2['Date'] = pd.to_datetime(df2['Date'], format='%d-%m-%Y', errors='coerce')


```

# EXPLORATORY DATA ANALYSIS

## Analyze Unemployment Trends

### 1. Visualize the unemployment rate over time for both datasets:

```python
# Plot unemployment rate over time for Dataset 1
plt.figure(figsize=(12, 6))
plt.plot(df1['Date'], df1['Estimated Unemployment Rate (%)'], marker='o', linestyle='-', label='Dataset 1')
plt.title('Estimated Unemployment Rate Over Time (Dataset 1)')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
plt.legend()
plt.show()

# Plot unemployment rate over time for Dataset 2
plt.figure(figsize=(12, 6))
plt.plot(df2['Date'], df2['Estimated Unemployment Rate (%)'], marker='o', linestyle='-', color='orange', label='Dataset 2')
plt.title('Estimated Unemployment Rate Over Time (Dataset 2)')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
plt.legend()
plt.show()
```

**General Trend** (dataset1)

The unemployment rate appears relatively stable between mid-2019 and early 2020.
However, there's a sharp spike around March–May 2020, likely due to the economic impact of COVID-19 lockdowns.
After May 2020, the unemployment rate gradually decreases but remains unstable.
Variation Across Categories:

Each line represents a different category (e.g., country, state, industry, or demographic group).
Before the spike, unemployment rates differ, but most values are between 0-30%.
During the peak, some categories reached over 70% unemployment.
Impact of COVID-19:

The surge in unemployment coincides with global lockdowns and business shutdowns during the early months of the pandemic.
Some categories were affected more severely than others.

**Interpretation of the Unemployment Rate Over Time (Dataset 2)**

Trend Overview:

The unemployment rate was relatively stable before April 2020.
There is a sharp peak in May 2020, where some categories reached over 70% unemployment.
After May, the unemployment rate declined significantly, but fluctuations continued.
Possible Causes of the Peak:

The COVID-19 pandemic led to widespread job losses and economic shutdowns.
The peak around May 2020 aligns with global lockdowns.
Post-Peak Trends:

The unemployment rate declined after May 2020, indicating economic recovery.
However, some groups remained above pre-pandemic levels, suggesting unequal recovery.
Differences from Dataset 1:

While Dataset 1 covered 2019-2020, this chart focuses on 2020 only.
Both datasets show a similar spike in unemployment around May 2020.

 ## 2.Analyze Unemployment by Region



```python
# Group by region and calculate average unemployment rate for Dataset 1
region_unemployment_1 = df1.groupby('Region')['Estimated Unemployment Rate (%)'].mean().reset_index()

# Plot average unemployment rate by region for Dataset 1
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=region_unemployment_1)
plt.title('Average Unemployment Rate by Region (Dataset 1)')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()

# Group by region and calculate average unemployment rate for Dataset 2
region_unemployment_2 = df2.groupby('Region')['Estimated Unemployment Rate (%)'].mean().reset_index()

# Plot average unemployment rate by region for Dataset 2
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=region_unemployment_2)
plt.title('Average Unemployment Rate by Region (Dataset 2)')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

**Comparison of Unemployment Rate by Region (Dataset 1 vs. Dataset 2)**

Similar Trends Across Datasets:

Haryana has the highest unemployment rate in both datasets (~26-27%).

Tripura also consistently shows a high unemployment rate.

Key Differences Between Dataset 1 and Dataset 2:

Dataset 1: Higher unemployment in Tripura compared to other states.

Dataset 2: Haryana remains the highest, but Punjab and Uttar Pradesh show higher unemployment.

**Possible Reasons for Differences:**

Different time periods or sample sizes used in each dataset.
Economic factors such as industry closures, policy changes, or survey methodology variations.

**Interpretation of Average Unemployment Rate by Region (Dataset 2)**

Regional Variations in Unemployment:

Some regions exhibit significantly higher unemployment rates than others.
Haryana has the highest unemployment rate, exceeding 27%.
Uttar Pradesh and Punjab also have relatively high unemployment rates.
Regions with Lower Unemployment:

States like Meghalaya, Kerala, and Odisha show lower unemployment rates.
This could indicate stronger local economies or alternative employment sources.
Possible Causes of Variation:

Industrial and Economic Factors: States with industrial hubs may have higher employment stability.

Agricultural Dependence: Rural and agrarian regions might experience seasonal employment variations.

Impact of COVID-19: Some states may have been more affected by lockdowns and business closures than others.
